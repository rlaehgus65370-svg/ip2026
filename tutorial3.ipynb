Dictionaries

```python
d = {'user':'bozo', 'pswd':1234}
print(d['user'])

d = {'user':'bozo', 'pswd':1234}
print(d['pswd'])

print(d['bozo'])
```

```python
d = {'user':'bozo', 'pswd':1234}
d['user'] = 'clown'
print(d)
```

```python
d = {'user':'bozo', 'pswd':1234}
d['id'] = 45
print(d)
```

```python
d = {'user':'bozo', 'p':1234, 'i':34}
del d['user']
print(d)

d = {'user':'bozo', 'p':1234, 'i':34}
d.clear()
print(d)

d = {'user':'bozo', 'p':1234, 'i':34}
a=[1,2]
del a[1]
print(a)
```

```python
d = {'user':'bozo', 'p':1234, 'i':34}
d.keys()

d = {'user':'bozo', 'p':1234, 'i':34}
d.values()

d = {'user':'bozo', 'p':1234, 'i':34}
d.items()
```
