```python
import cv2
import numpy as np
```

```python
def nothing(x):
    pass
```

```python
drawing=False#trueifmouseispressed
ix,iy=-1,-1
cx,cy=-1,-1
org_img = cv2.imread('tkakr.jpg')   
img = org_img.copy()
def draw_circle(event,x,y,flags,param):
    global ix,iy,cx,cy,drawing,mode,img

    if event==cv2.EVENT_LBUTTONDOWN:
        drawing=True
        ix,iy=x,y
    elif event==cv2.EVENT_MOUSEMOVE:
        cx,cy = x,y   
            
    elif event==cv2.EVENT_LBUTTONUP:
        ix,iy=-1,-1 
        cx,cy=-1,-1 
        drawing=False
```

```python
cv2.namedWindow('image')
cv2.createTrackbar('value','image',0,255,nothing)
cv2.setMouseCallback('image', draw_circle)
font = cv2.FONT_HERSHEY_SIMPLEX
while(1):
    img = org_img.copy()
    value = cv2.getTrackbarPos('value','image')
    if drawing == True:
        cv2.rectangle(img,(ix,iy),(cx,cy),(0,255,0),-1)
    #cv2.putText(img,f"tkakr position : ({ix}',{iy})-({cx},{cy})-{value}",(10,30), font, 0,4(255,255,255),1,cv2.LINE_AA)
    cv2.imshow('image',img) 
    k=cv2.waitKey(1)&0xff   
    if k==27:
        break       
cv2.destroyAllWindows()
```
